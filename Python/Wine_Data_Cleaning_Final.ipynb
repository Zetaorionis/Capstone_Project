```python
# Creating categories based on the Wine Spectator's 100-point Scale (https://www.wineenthusiast.com/wine-ratings/)
# based on the Wine Enthusiast Rating Scale:

# 98-100
# Classic
# The Pinnacle of quality

# 94-97
# Superb
# A great achievement

# 90-93
# Excellent
# Highly recommended

# 87-89
# Very Good
# Often good value; well recommended

# 83-86
# Good
# Suitable for everyday consumption, often good value

# 80-82
# Acceptable
# Can be employed

# Define bin edges and labels
bins = [80, 83, 87, 90, 94, 98, 100]  # defines 6 bins
short_labels = [
    'Acceptable',
    'Good',
    'Very Good',
    'Excellent',
    'Superb',
    'Classic'
]  # 6 labels

# Ensure the number of labels is one fewer than the number of bins
assert len(short_labels) == len(bins) - 1, "Labels must be one fewer than the number of bins."

# Create the 'rating_category' column with proper bins and labels
wine_df['rating_category'] = pd.cut(wine_df['points'], bins=bins, labels=short_labels, include_lowest=True)

# Check the new DataFrame to ensure proper categorization
# print(wine_df).head()  # Display the first few rows
```

### Add "vintage" column/feature via Regex Extraction

```python
# Import Regex library 
import re
```

```python
#Create copy of winery column for Regex string
clean_wine_data_final['winery_copy'] = clean_wine_data_final['winery']
```

```python
# Escape any special characters in winery string
clean_wine_data_final['winery_copy'] = clean_wine_data_final['winery_copy'].apply(lambda x : re.escape(x))
```

```python
# Fine sample wineries with special character '(' and check escaping 
p = '\('
clean_wine_data_final[['winery', 'winery_copy']][clean_wine_data_final['winery'].str.contains(p)]
```

```python
# Define function that will extract the vintage/year usig regex

## Don't need >1 conditional becasue there are only 2 wines with more than one and the first one is the accurate vintage(manually confirmed) 

def extract_vintage(title,winery):
    vintage = re.findall('(?:'+ winery +'\D*)([\d+]{4})',title)
    if len(vintage) == 0: 
        return 'No Year'
    #if len(vintage) > 1: 
    #    return 'Multiple'
    else: 
        return vintage[0]
```

```python
# Apply extract_vintage function to create vintage column, passing title and escaped winery text
clean_wine_data_final['vintage'] = clean_wine_data_final.apply(lambda x: extract_vintage(x['title'], x['winery_copy']), axis=1)
```
